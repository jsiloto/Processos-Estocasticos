# Solução do Projeto

## 1 Introdução

```python
h, samplerate = librosa.load('Garrafa.wav')
h /= np.max(np.abs(h),axis=0)
h = h[:6000]
```

```python
def make_measurement(h, num_samples=20000):
    # Simulate recording of impulse response 
    # h[k]: Real impulse response
    # w[k]: White noise used to excite impulse response 
    # n[k]: White measurement noise (var=0.01)
    # r[k] = w[k] ∗ h[k] + n[k].
    w_var = 1.0
    n_var=0.01
        
    w = np.random.normal(0, np.sqrt(w_var), size=num_samples)
    n = np.random.normal(0, np.sqrt(n_var), size=num_samples)
    r = scipy.signal.convolve(w, h) 
    return w, r
```

```python
def estimation_error(estimation, original):
    return np.sum((estimation - original)**2)/np.sum(original**2)
```

```python
def show_results(ro, h):
    error = estimation_error(ro, h)
    print("Estimation error = {}".format(error))
    plt.plot(h)
    plt.plot(ro, alpha=0.7)
    plt.show()
    
```

## 2 Estimação por Correlação

![title](q3.jpeg)

```python
def correlation_estimator(h, num_samples=1000):
    w, r = make_measurement(h, num_samples)
    ro = np.convolve(r, w[::-1])/len(w)
    ro = ro[num_samples-1:num_samples+len(h)-1]
    return ro
```

```python
#METHOD 1
num_samples=100000
ro = correlation_estimator(h, num_samples)
method1_error = show_results(ro, h)
```

## 3 Melhorando a Correlação

```python
#METHOD 2 - average of multiple shorter samples
ro_list = []
num_samples = 1000
num_measures = 100
for i in range(num_measures):
    ro = correlation_estimator(h, num_samples)
    ro_list.append(ro)
    
ro = sum(ro_list)/len(ro_list)
method2_error = show_results(ro, h)
```

```python
#METHOD 2.5 - Chop up large measure into smaller ones
num_samples=100000
sub_sample=len(h)
indices = np.arange(sub_sample, num_samples, sub_sample)
w, r = make_measurement(h, num_samples)
r_list = np.split(r, indices)
w_list = np.split(w, indices)
ro_list = []
for i in range(len(r_list)-1):
    ro = np.convolve(r_list[i], w_list[i][::-1])/len(w_list[i])
    ro = ro[sub_sample-1:]
    ro = np.pad(ro, (0, len(h)-len(ro)), 'constant')
    ro_list.append(ro)
    
ro = sum(ro_list)/len(ro_list)
method25_error = show_results(ro, h)
```
